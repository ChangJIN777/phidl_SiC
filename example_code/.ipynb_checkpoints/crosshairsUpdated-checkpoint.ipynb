```python
from phidl import Device
from phidl import quickplot as qp # Rename "quickplot()" to the easier "qp()"
import phidl.geometry as pg
import numpy as np
```

```python
D = Device()
xpts1 = (0, 1, 1, 1, 2)
ypts1 = (0, 1, 0, -1, 0)
poly1 = D.add_polygon( [xpts1, ypts1], layer = 0)

xpts2 = (0+20, 1+20, 1+20, 1+20, 2+20)
ypts2 = (0, 1, 0, -1, 0)
poly2 = D.add_polygon( [xpts2, ypts2], layer = 0)


xpts3 = (0+20, 1+20, 1+20, 1+20, 2+20)
ypts3 = (0+20, 1+20, 0+20, -1+20, 0+20)
poly3 = D.add_polygon( [xpts3, ypts3], layer = 0)

xpts4 = (0, 1, 1, 1, 2)
ypts4 = (0+20, 1+20, 0+20, -1+20, 0+20)
poly4 = D.add_polygon( [xpts4, ypts4], layer = 0)

qp(D)
```

```python
bigSq = Device()
        
smallSqRef = bigSq.add_array(D, columns = 16, rows = 16, spacing = [60, 60])
    
qp(bigSq)
```

```python
horiz = []
for i in range(1, 17):
    T = pg.text(text = str(i), size = 5)
    horiz.append(T)
    
H = pg.grid(horiz,
            spacing = (56,1),
            separation = True,
            shape = (60,1),
            align_x = 'x',
            align_y = 'y',
            edge_x = 'x',
            edge_y = 'ymax')

H.movey(-30)

qp(H)
```

```python
vert = []
for i in range(1, 17):
    T = pg.text(text = str(i), size = 5)
    vert.append(T)
    
vert.reverse()
    
V = pg.grid(vert,
            spacing = (1,54),
            separation = True,
            shape = (1,20),
            align_x = 'x',
            align_y = 'y',
            edge_x = 'x',
            edge_y = 'ymax')

V.center = [0, 460]
V.movex(-30)

qp(V)
```

```python
bigSq << H # this adds a reference to G to the original bigSq
bigSq << V

qp(bigSq)
```

```python
totalDev = Device()
totalDev1 = totalDev.add_array(bigSq, columns = 2, rows = 2, spacing = [2000, 2000])

bigNum = []
for i in range(1, 5):
    T = pg.text(text = str(i), size = 10)
    bigNum.append(T)
    
Num = pg.grid(bigNum,
            spacing = (2000, 2000),
            separation = True,
            shape = (2,2),
            align_x = 'x',
            align_y = 'y',
            edge_x = 'x',
            edge_y = 'ymax')

Num.movex(-50)
Num.movey(-50)

totalDev << Num
totalDev.center = [0, 0]

qp(totalDev)
```

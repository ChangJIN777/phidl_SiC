## Import the geometry from existing GDS files

```python
DIR = '/Users/changjin/Desktop/Research/waveguide_solver/phidl_SiC/example_code/SiC_cavity_v1/'
file_name = 'SiC_cavity_v1_num_0.gds'

# importing the geometry from existing GDS files 
D = Device() #create blank device and add unit cell to it
c = pg.import_gds(DIR+file_name, cellname = None, flatten = False)
qp(c)
```

```python
# invert the geometry (for negative resists)
ic = pg.invert(c,border=0)
qp(ic)
h = c.ymax-c.ymin
w = c.xmax-c.xmin
print(h) # for debugging purpose
print(w) # for debugging purpose
```

```python
# adding the necessary ports for connecting them with other geometries
c.add_port(name='N', midpoint= [0,h/2], width = w, orientation = 90)
c.add_port(name='S', midpoint= [0,c.ymin], width = w, orientation = 270)
trenchT = D << pg.compass([10,500e-3]) #define rectangular trench using compass labels for ports
trenchB = D << pg.compass([10,500e-3]) #define rectangular trench using compass labels for ports
c = D << c # adding the geometry c to device D 

c.connect(port = 'N', destination = trenchT.ports['S']) #stack trench-waveguide-trench for unit cell
trenchB.connect(port = 'N', destination = c.ports['S'])

D.center = [0,0] #move device to origin
qp(D) # cavity with the trench objects attached 
```

```python
# removing the cavity, leaving only the trenches and the bridge in between
D.remove(c)
qp(D)
# adding the holes to the bridge in between 
ic = D << ic
qp(D)
```

```python
mystyle = {(0.01, 0): {'fill': '#CC00FF','stroke': 'black'}}
D.write_svg('Nanobeam.svg', style=mystyle)

obj_width = D.ymax-D.ymin #calculate total height of cell
print(obj_width)

pad = Device()
pad_ref = pad.add_array(D, columns = 1, rows = 6, spacing = [0,obj_width])
pad.center = [0,0]

# add label to pad
label = Device()
T = pg.text('1', layer = 0, size = 0.5)
label << pg.invert(T, border = 0.15)
label.align(elements = 'all', alignment = 'x')
label.align(elements = 'all', alignment = 'y')
label.center = [0,0]
label.movex(pad.xmin-0.5)
label.movey(pad.ymin-0.5)
pad << label
qp(pad)

# add pad to full design
samples.append(pad)
```

## Testing the make cavity function 

```python
DIR = '/Users/changjin/Desktop/Research/waveguide_solver/phidl_SiC/example_code/SiC_cavity_v1/'
file_name = 'SiC_cavity_v1_num_0.gds'
fullfilepath = DIR+file_name
dev = make_cavity_v2(fullfilepath)
qp(dev)
```

```python
 # import the device from the desired gds file 
cav = import_GDS(fullfilepath)
# get the height and the width of the imported device 
h = cav.ymax-cav.ymin
w = cav.xmax-cav.xmin
xR = cav.xmax
xL = cav.xmin
cav.add_port('cav_wvg_port',midpoint=(xR,0),width = h, orientation= 0)
cav.add_port('cav_wvg_port_left',midpoint=(xL,0),width = h, orientation= 180)

trenchW = 6
wvg_length = 5
teth_length = 10
teth_width = 1.3*h

wvg_section = pg.rectangle((wvg_length,h))
wvg_section.add_port('wvg_left',(0,h/2), orientation= 180)
wvg_section.add_port('wvg_right',(wvg_length,h/2), orientation= 0)
qp(wvg_section)
```

```python
# make the tether region 
tether = make_tether_2(trenchW,teth_length,teth_width,h)
qp(tether)
```

```python
# make the tap region 
tap_wvg = pg.taper(30,width1=h,width2=.06)
tap_wvg.add_port('tap_wvg_left',(0,0), orientation= 180)
qp(tap_wvg)
```

```python
E = Device('full_device')
cav_ref = E.add_ref(cav)
wvg_ref1 = E.add_ref(wvg_section)
wvg_ref1.connect('wvg_left', destination=cav_ref.ports['cav_wvg_port'])
teth_ref1 = E.add_ref(tether)
teth_ref1.connect('tether_left', destination=wvg_ref1.ports['wvg_right'])
tap_ref = E.add_ref(tap_wvg)
tap_ref.connect('tap_wvg_left', destination=teth_ref1.ports['tether_right'])
qp(E)
```

```python
# adding the trench area 
full_dev_width = w + teth_length+wvg_length + 30 + 10
trench_rect = pg.rectangle((full_dev_width,trenchW)).move((xL,-trenchW/2))

qp(trench_rect)
```

```python
# adding everything together 
dev = pg.xor_diff(trench_rect,E)
qp(dev)
```

```python
# write the GDS 
dev.write_gds('test_device_v1.gds')
```
